```python
dtst.fillna(dtst.mean(), inplace=True)
dtst.isnull().any()
```

```python
A = dtst.iloc[:,1:-1]
B = dtst.iloc[:,-1]
print(A.shape,B.shape)
```

1.a) Applying PCA on CC dataset.


```python
pca = PCA(3)
a_pca = pca.fit_transform(A)
principalDf = pd.DataFrame(data = a_pca, columns = ['principal component 1', 'principal component 2', 'principal component 3'])
finalDf = pd.concat([principalDf, dtst.iloc[:,-1]], axis = 1)
finalDf.head()
```

```python
#1.b Apply K Means on PCA Result
A = finalDf.iloc[:,0:-1]
B = finalDf.iloc[:,-1]
```

```python
ncluster = 3
#this is k in kmeans
km = KMeans(n_clusters=ncluster)
km.fit(A)

B_cluster_kmeans = km.predict(A)
#predicting the cluster for each data point

print(classification_report(B,B_cluster_kmeans, zero_division=1))
print(confusion_matrix(B, B_cluster_kmeans))
#summary of the predictions made by the classifier

train_accuracy = accuracy_score(B, B_cluster_kmeans)
print("\nAccuracy for our Training dataset with PCA:", train_accuracy)

score = metrics.silhouette_score(A, B_cluster_kmeans)
print("Sihouette Score: ",score) 
#Calculated sihouette Score
```

Sihouette Score always ranges from −1 to +1
High value indicates that the object is well matched to its own cluster and poorly matched to neighboring clusters.

```python
#predicting the cluster for each testing data point
B_clus_test = km.predict(A_test)

# Summary of the predictions made by the classifier
print(classification_report(B_test, B_clus_test, zero_division=1))
print(confusion_matrix(B_test, B_clus_test))

train_accuracy = accuracy_score(B_test, B_clus_test)
print("\nAccuracy of Training dataset with PCA:", train_accuracy)

#Calculate sihouette Score
score = metrics.silhouette_score(A_test, B_clus_test)
print("Sihouette Score: ",score) 

"""
Sihouette Score- ranges from −1 to +1 , a high value indicates that the object is well matched to its own cluster and poorly matched to neighboring clusters.
"""
```

```python
#Difference between PCA and LDA
"""Both LDA and PCA rely mainly on linear transformations and it's main aim to maximize the variance in lower dimension. PCA is however a unsupervised learning algorithm where LDA is a supervised learning algorithm. It means that PCA finds directions of maximum variance regardless of class labels while LDA finds directions of maximum class separability"""
```

```python
#PCA
"""It reduces features into smaller subset of orthogonal variables, called principal components – linear combinations of the original variables. The first component captures the largest variability of the data, while the second captures the second largest, and so on."""
```

```python
#LDA
"""LDA finds the linear discriminants in order to maximize the variance between the different categories while minimizing the variance within the class."""
```
